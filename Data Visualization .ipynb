```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Real World Examples

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.figure(figsize = (15,10))
plt.title("Gas Price Comparison in USD", fontdict={'fontweight':'bold','fontsize':30})
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year , gas[country] , marker = '.' , label = country)
# plt.plot(gas.Year , gas.USA , 'b.-', label='United States')
# plt.plot(gas.Year , gas.Canada , 'r.-', label = 'Canada' )
# plt.plot(gas.Year, gas["France"], 'g.-', label = 'France')
plt.legend()
plt.xlabel("Year")
plt.ylabel("USD")
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.show()
```

```python
fifa = pd.read_csv("fifa_data.csv")
fifa
```

```python

bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins = bins)
plt.title("Distribution of Player Skills in Fifa")
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")
plt.xticks(bins)
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

label = ['Left', 'Right']

colors = ['#abcdef','#aaffcc']
plt.pie([left, right], labels = label, colors= colors , autopct = "%.2f")
```

```python
fifa.Weight = [x.strip('lbs') if type(x) == str else x for x in fifa.Weight]
```

```python
fifa.Weight = fifa.Weight.dropna()
fifa.Weight = fifa.Weight.astype('float')
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight <150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight>200].count()[0]
```

```python
weights = [light, light_medium, medium,medium_heavy,heavy]
labels=['Under 125','125-150','150-175','175-200','>200']

explode = (.4,.2,0,0,.4)
plt.pie(weights,labels=labels, autopct = '%.2f %%', pctdistance=0.8,explode=explode)
plt.title("Weight Distributions of FIFA Players (in lbs)")
plt.style.use('ggplot')
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
plt.style.use("default")
labels= ['FC Barcelona', 'FC Madrid']
boxes = plt.boxplot([barcelona, madrid], labels = labels , patch_artist= True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # set edge color
    box.set(color='#4286f4')
    box.set(facecolor='#e0e0e0')
    
plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA Overal Rating")
plt.show()
```
